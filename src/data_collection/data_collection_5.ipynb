```python
import pandas as pd
import requests
```

```python

def fetch_usgs_turbine_locations(url="https://eersc.usgs.gov/api/uswtdb/v1/turbines"):
    try:
        data = pd.read_json(url)
        return data
    except Exception as e:
        print(f"Error fetching USGS turbine data: {e}")
        return pd.DataFrame()
x = fetch_usgs_turbine_locations()
x.head()
```

```python
df = x.copy()
```

```python
df.columns
```

```python
df = df[['case_id','t_state','p_name','xlong', 'ylat', 'p_cap', 'p_year', 't_manu', 't_model', 'eia_id']]
df.rename(columns={
    't_state': 'state',
    'p_name': 'plant_name',
    'xlong': 'longitude',
    'ylat': 'latitude',
    'p_cap': 'capacity_MW',
    'p_year': 'year_operational',
    't_manu': 'manufacturer',
    't_model': 'model',
    'eia_id': 'plantcode'
}, inplace=True)
df.head()
```

```python
df.info()
```

```python
df = df[df['state'] == 'TX']
df.info()
```

```python
df.reset_index(inplace=True)
```
