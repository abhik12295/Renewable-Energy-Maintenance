```python
# Initialize the NCEI Bot with the token
ncei_bot = NCEIBot(TOKEN_NCEI, cache_name="ncei_cache")
```

```python

# fetching all weather data for country united states
dataset_id = "GHCND"  # Global Historical Climatology Network Daily
location_id = "FIPS:48"  # FIPS code for Texas (48)
mindate = date(2021, 1, 1)  # Start date
maxdate = date(2024, 12, 31)  # End date
datatype_ids = ["TMAX", "TMIN", "PRCP", "AWND"]  # Maximum temperature, minimum temperature, and precipitation
```

```python
# get all the stations from Texas
stations = ncei_bot.get_stations(
    datasetid=dataset_id,
    datatypeid = datatype_ids,
    locationid=location_id,
    startdate=mindate,
    enddate=maxdate
)
```

```python
print(f"Found {len(stations)} stations in Texas.")
# select the station with best coverage
```

```python
station = sorted(stations.values(), key=lambda s: -int(s["datacoverage"]))[0]
print(station)
```

```python
# convert station to dict to dataframe
import pandas as pd
df_station = pd.DataFrame([station])
df_station.head()
```

```python
### testing

ncei_bot1 = NCEIBot(TOKEN_NCEI, cache_name="ncei_cache1")
mindate = date(2025, 1, 1)
maxdate = date(2025, 6, 30)
final_res = []
#df_stations_texas_id = ['GHCND:USC00410058', 'GHCND:USC00412114']
for station_id in df_stations_texas_id:
    yearly_res = []
    year = maxdate.year
    while year >= mindate.year:
        response_year = ncei_bot1.get_data(
            datasetid=dataset_id,
            stationid=station_id,
            datatypeid=datatype_ids,
            startdate=date(year, 1, 1),
            enddate=date(year, 12, 31)
        )
        yearly_res.append(list(response_year.values()))
        year -= 1
    final_res.extend(yearly_res)
```

```python
ncei_bot1 = NCEIBot(TOKEN_NCEI, cache_name="ncei_cache1")
mindate = date(2015, 1, 1)
maxdate = date(2025, 12, 31)
final_res = []
for station_id in df_stations_texas_id:
    yearly_res = []
    year = maxdate.year
    while year >= mindate.year:
        response_year = ncei_bot1.get_data(
            datasetid=dataset_id,
            stationid=station_id,
            datatypeid=datatype_ids,
            startdate=date(year, 1, 1),
            enddate=date(year, 12, 31)
        )
        yearly_res.append(list(response_year.values()))
        year -= 1
    final_res.extend(yearly_res)

```

```python
df_final_texas_all = pd.DataFrame()
for i in final_res:
    df_final_texas_all = pd.concat([df_final_texas_all, pd.DataFrame(i)], ignore_index=True)
```

```python
# Fix: aggregate duplicates before unstacking to avoid ValueError
df_final_texas_all_sorted = df_final_texas_all.sort_values(['date', 'station', 'datatype'])
# Group by and aggregate (e.g., take the mean if duplicates exist)
df_grouped = df_final_texas_all_sorted.groupby(['date', 'station', 'datatype'], as_index=False)['value'].mean()
df_final_texas_all_pivot = df_grouped.pivot(index=['date', 'station'], columns='datatype', values='value').reset_index()
# Now df_final_texas_all_pivot contains all date-station combinations with columns for each datatype

df_final_texas_all_pivot
```
